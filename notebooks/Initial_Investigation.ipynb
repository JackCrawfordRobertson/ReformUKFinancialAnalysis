```python
import pandas as pd

# Assuming there's a data file in the 'data' directory
data = pd.read_csv('../data/Reform_UK_results.csv')
print(data.head())
print(data.columns)
```

```python
# Display basic information about the DataFrame
print(data.info())

# Show summary statistics for numerical columns
print(data.describe())

# Display the column names
print(data.columns)

# Check for missing values
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart for the number of donations by regulated entity
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='RegulatedEntityName', order=data['RegulatedEntityName'].value_counts().index)
plt.xticks(rotation=0,  )
plt.title('Number of Donations by Regulated Entity')
plt.xlabel('Regulated Entity')
plt.ylabel('Count')
plt.show()


```
